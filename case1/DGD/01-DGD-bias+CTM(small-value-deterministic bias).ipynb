IWDGD中小值的确定性噪声 对应到DGD算法中是固定的对偶攻击值为-140.7 这样两个算法各自迭代1000次 所有用户的对偶迭代值均在140.2附近

鲁棒聚合器CTM 每个用户剔除3个用户对偶信息（包含自身的信息）3-是网络中噪声用户数量的上界

```python
#包文件
import numpy as np
import matplotlib.pyplot as plt
import random
from sympy import *
import math 
from datetime import datetime
```

```python
#定义变量 
agent_resource_quantity = np.zeros(shape=(100,1000)) #100个agent的资源量 
dual_variable = np.zeros(shape=(100,1000))#对偶变量
sum_resource=np.zeros(shape=(1,1000), dtype=np.float64)#存放每一次迭代后所有用户资源量的总和
sum_cost=np.zeros(shape=(1,1000), dtype=np.float64)#存放每一次迭代后所有用户成本的总和
iteration_update = 0.01#迭代步长
A=np.loadtxt("100_Network_Topology_Double_Random_Adjacency_Matrix_Degree_15.txt") #网络拓扑的邻接矩阵 对角元素要为0
W=np.zeros(shape=(100,100)) #双随机的网络拓扑矩阵W（对角线处非零）
iteration=[]#绘图横坐标
a=np.loadtxt("100_uniform_distribution_(1,2)_point1.txt")#每个agent的成本函数中的超参数
b=np.loadtxt("100_normal_distribution_(2,0.6)_point1.txt")#每个agent的成本函数中的超参数
q=3 #TM鲁棒聚合中丢弃的数据数量 
neighbor_help=[]
```

```python
#根据邻接矩阵生成网络对应的双随机的权重混合矩阵
for i in range(0,100):
    for j in range(0,100):
        if A[i][j]!=0 or i==j:
            W[i][j]=1/16

#对偶变量的初始化 初始化一致
for i in range (0,100):
    dual_variable[i][0]=0
    
#资源量的初始化 资源量的初始化需要和对偶变量的初始值满足某种关系
for i in range (0,100):
    agent_resource_quantity[i][0]=50

iteration.append(0)

#总资源量 初始化
sum=0
for i in range (0,100):
    sum=sum+agent_resource_quantity[i][0]
sum_resource[0]=sum

#用户总成本初始化
sum=0
for i in range(0,100):
    sum=sum+a[i]*(agent_resource_quantity[i][0]-b[i])*(agent_resource_quantity[i][0]-b[i])
sum_cost[0][0]=sum

#对每个用户的15个邻居进行初始化
for i in range (0,100):
    for j in range (0,100):
        if A[i][j]!=0:
            neighbor_help.append(j)
neighbor=np.array(neighbor_help).reshape(100,15)
```

```python
#DGD算法 
def DGD_CTM_deterministic_bias(a,b,agent_resource_quantity,sum_resource,iteration_update,W,iteration,neighbor):
    #显示初始化的数据
    #输出初始状态的原变量和对偶变量
    print("****************************DGD算法+计算误差+CTM****************************")
    print("相关变量初始值为：")
    for i in range (0,100):
        print("agent",(i+1),"的初始资源申请量为:",agent_resource_quantity[i][0])
    print("*************************")
    print("初始资源申请量总和为:",sum_resource[0][0])
    print("*************************")
    print("初始成本总和为:",sum_cost[0][0])
    print("*************************")
    for i in range(0,100):
        print("agent",(i+1),"的初始对偶变量为:",dual_variable[i][0])
    print("*************************")
    print("迭代步长为:",iteration_update)
    print("*****************************************************************************")
    
    #开始迭代
    for k in range(1,1000):#迭代次数
        iteration.append(k)
        
        #所有用户对偶变量的迭代 
        #阶段一：梯度下降
        dual_variable_intermediate=[] #存放当前迭代中100个用户对对偶变量做完梯度下降后得到的中间值
        for i in range(0,100):
            if i==7 or i==20 or i==34:
                dual_variable_intermediate.append(-140.7)
            else:
                dual_variable_intermediate.append(dual_variable[i][k-1]-iteration_update*(50-agent_resource_quantity[i][k-1]))       
        
        #阶段二：鲁棒聚合 每个用户用鲁棒聚合器CTM聚合自身以及邻居的对偶值   
        for i in range (0,100):
            mix_sum_dual=0
            sort_list=[]
            sum=0
            for j in range (0,15):
                sort_list.append(dual_variable_intermediate[neighbor[i,j]])
            sort_list.append(dual_variable_intermediate[i])
            sort_list.sort()
            #剔除最大的q个数据和最小的q个数据
            for j in range(0,int(q)):
                del sort_list[15-j]
            for j in range(0,int(q)):
                del sort_list[0]     
            for j in range(0,len(sort_list)):
                mix_sum_dual=mix_sum_dual+sort_list[j]
            #用户取剩下数据的平均
            dual_variable[i][k]=(1/(16-2*q))*(mix_sum_dual)
        
        #用户资源申请量的迭代
        for i in range (0,100):
            agent_resource_quantity[i][k]=(dual_variable[i][k]-2*a[i]*b[i])/(-2*a[i])

        #计算所有用户的资源量总和
        sum1=np.float64(0)
        for i in range (0,100):
            sum1 = sum1 + agent_resource_quantity[i][k]
        sum_resource[0][k]=sum1
        
        #计算所有用户的成本总和
        sum2=np.float64(0)
        for i in range(0,100):
            sum2=sum2+a[i]*(agent_resource_quantity[i][k]-b[i])*(agent_resource_quantity[i][k]-b[i])
        sum_cost[0][k]=sum2
        
        #输出第k次迭代中的结果 每间隔2000次迭代输出一次结果
        if (k+1)%50==0:
        #if (k+1)%2000==0:
            print(k,"次迭代中的结果：")
            for i in range (0,100):
                print("agent",(i+1),"在",k,"次迭代中的资源申请量为:",agent_resource_quantity[i][k])
            print("资源申请量总和为:",sum_resource[0][k])
            print("****************************************************************************")
            print("成本总和为:",sum_cost[0][k])
            print("****************************************************************************")
            for i in range(0,100):
                print("agent",(i+1),"在",k,"次迭代中的对偶变量为:",dual_variable[i][k])   
            print("*************************")
            print("迭代步长为:",iteration_update)
            print("****************************************************************************")
        
        iteration_update=1/(pow(k+1,0.6)) #递减步长
        
    print(datetime.now().strftime('%Y-%m-%d %H:%M:%S.%f')[:-3])
    
    return None
```

```python
DGD_CTM_deterministic_bias(a,b,agent_resource_quantity,sum_resource,iteration_update,W,iteration,neighbor)
```
