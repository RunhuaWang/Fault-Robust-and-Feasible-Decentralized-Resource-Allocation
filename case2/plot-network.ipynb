```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import os

# 
A = np.loadtxt("Adjacency_Matrix_generator_zero.txt")
G = nx.from_numpy_array(A)

#  1  N
labels = {i: str(i + 1) for i in G.nodes()}

# 
pos = nx.spring_layout(G, seed=42, k=13 / np.sqrt(len(G.nodes)))

# 
node_color = 'lightblue'
node_size = 900
font_size = 16
font_color = 'black'
font_family = 'Times New Roman'
font_weight = 'bold'

# 
plt.figure(figsize=(10, 10))

# 
nx.draw(
    G,
    pos,
    with_labels=False,
    node_size=node_size,
    node_color=node_color,
    edge_color='gray',
)

# 1  N
nx.draw_networkx_labels(
    G,
    pos,
    labels=labels,
    font_size=font_size,
    font_color=font_color,
    font_family=font_family,
    font_weight=font_weight,
)

# 
os.makedirs("./figures", exist_ok=True)

# 
plt.axis('off')
plt.savefig('./figures/network.pdf', format='pdf', bbox_inches='tight')
plt.savefig('./figures/network.png', format='png', bbox_inches='tight')
plt.show()
```
