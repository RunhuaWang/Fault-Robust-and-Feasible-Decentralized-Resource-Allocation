小值高斯分布

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sympy import *
import math
```

```python
mu1 = 60  #期望
sigma1 = 10  #标准差
num1 = 2000  #个数
```

```python
random_bias_normal_distribution1 = np.random.normal(mu1, sigma1, num1)
```

```python
random_bias_normal_distribution1 = np.round(random_bias_normal_distribution1, decimals=1)
```

```python
random_bias_normal_distribution1
```

```python
min(random_bias_normal_distribution1)
```

```python
max(random_bias_normal_distribution1)
```

```python
np.savetxt('2000_Gaussian_distribution_(60,10).txt',random_bias_normal_distribution1,fmt='%0.8f')#数据存储
```

```python
beta1= np.loadtxt("2000_normal_distribution_(60,10).txt")
```
